### A claim shows that average total bill of new jersey  is $22.


```python
bill=df.total_bill
```

```python
Ho :: Avg total bill = 22
        
Ha :: Avg =! 22
        
confidence level = 95

significance = 5

p critical= 0.05

```

taking sample

```python
bill
```

Choosing the t test as we have only population  mean and std

```python
t_test=stats.ttest_1samp(bill,22)
```

```python
p_value=t_test[1]
```

```python
p_value
```

```python
p_value > 0.05
```

Since the p_value is lesser than p critical, we reject the null hypothesis that avg of total bill
is 22. In other manner we can say that avg of total bill is not 22.

### Check weather in general male tip higher than female in new jersey area. We have the population sample mean and std

```python
df
```

```python
f_tip=df[df.sex=='Female'].tip
```

```python
m_tip=df[df.sex=='Male'].tip
```

```python
t_test=stats.ttest_ind(f_tip,m_tip)
```

```python
p_value=t_test[1]
```

```python
p_value >0.05
```

Since the p value is greater than p critical, we fail to reject the null hypothesis hence we can 
say that female tip is equal to male tip

### Is there any relationship between gender of a person and their smoking habit 

```python
df
```

```python
df1=df[['sex','smoker']]
```

```python
female=df1[df1.sex=='Female']
```

```python
male=df1[df1.sex=='Male']
```


Ha = There is a relationship between gender and smoking habit

Confidence level = 95

Significance level = 5

P critical = 0.05

We have the sample already.

Choosing the chi square test as we are finding the relationship between two categorical columns.





```python
obs=pd.crosstab(df1.sex,df1.smoker)
```

```python
chi_test=stats.chi2_contingency(obs)
```

```python
p_value=chi_test[1]
```

```python
p_value
```

```python
p_value >0.05
```
